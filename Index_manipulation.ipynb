In this notebook I arrange the index using a multiindex object to create a dataframe with multilevel index

```python
#Import necessary libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data Market.csv')
df.head()
```

We see that the headers do not make sense. We drop the first row and drop the first column as it is redundant with the second column, and use said second column as the index. Finally, we transpose the Dataframe for ease of constructing the multilevel index.

```python
df = pd.read_csv('Data Market.csv')
df = df.drop(0)
df.index = df.loc[:,'Unnamed: 1']
df.index.names = [df.index[1]]
df = df.drop(columns=['Unnamed: 0','Unnamed: 1']).T

df.head(10)
```

We then move the first six rows of the second column to the first, and we fill all the empty rows of the first column.

```python
df_index = df.iloc[:,0:2].ffill().bfill(axis=1)
df_index.iloc[:6,1]=' '
df_index.head(10)
```

Finally, we set `df_index` as the new index, we remove the decrepit columns, and we tranpose the Dataframe to return to its original orientation.

```python
mult_index = pd.MultiIndex.from_frame(df_index, names=['Name', 'Year'])
df_header = df.iloc[:,2:].set_index(mult_index,drop=True).T

df_header.head(10)
```
